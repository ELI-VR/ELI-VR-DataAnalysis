```python
for frame in FP_area_1:
    timestamps["fp_1"].append(frame['TimeStamp'])
    HMDPositionGlobal["fp_1"]["x"].append(frame['HMDPositionGlobal']["x"])
    HMDPositionGlobal["fp_1"]["y"].append(frame['HMDPositionGlobal']["y"])
    HMDPositionGlobal["fp_1"]["z"].append(frame['HMDPositionGlobal']["z"])
    HMDPositionLocal["fp_1"]["x"].append(frame['HMDPositionLocal']["x"])
    HMDPositionLocal["fp_1"]["y"].append(frame['HMDPositionLocal']["y"])
    HMDPositionLocal["fp_1"]["z"].append(frame['HMDPositionLocal']["z"])

for frame in FP_area_2:
    timestamps["fp_2"].append(frame['TimeStamp'])
    HMDPositionGlobal["fp_2"]["x"].append(frame['HMDPositionGlobal']["x"])
    HMDPositionGlobal["fp_2"]["y"].append(frame['HMDPositionGlobal']["y"])
    HMDPositionGlobal["fp_2"]["z"].append(frame['HMDPositionGlobal']["z"])
    HMDPositionLocal["fp_2"]["x"].append(frame['HMDPositionLocal']["x"])
    HMDPositionLocal["fp_2"]["y"].append(frame['HMDPositionLocal']["y"])
    HMDPositionLocal["fp_2"]["z"].append(frame['HMDPositionLocal']["z"])
    
for frame in FP_area_3:
    timestamps["fp_3"].append(frame['TimeStamp'])
    HMDPositionGlobal["fp_3"]["x"].append(frame['HMDPositionGlobal']["x"])
    HMDPositionGlobal["fp_3"]["y"].append(frame['HMDPositionGlobal']["y"])
    HMDPositionGlobal["fp_3"]["z"].append(frame['HMDPositionGlobal']["z"])
    HMDPositionLocal["fp_3"]["x"].append(frame['HMDPositionLocal']["x"])
    HMDPositionLocal["fp_3"]["y"].append(frame['HMDPositionLocal']["y"])
    HMDPositionLocal["fp_3"]["z"].append(frame['HMDPositionLocal']["z"])
    
for frame in FP_area_4:
    timestamps["fp_4"].append(frame['TimeStamp'])
    HMDPositionGlobal["fp_4"]["x"].append(frame['HMDPositionGlobal']["x"])
    HMDPositionGlobal["fp_4"]["y"].append(frame['HMDPositionGlobal']["y"])
    HMDPositionGlobal["fp_4"]["z"].append(frame['HMDPositionGlobal']["z"])
    HMDPositionLocal["fp_4"]["x"].append(frame['HMDPositionLocal']["x"])
    HMDPositionLocal["fp_4"]["y"].append(frame['HMDPositionLocal']["y"])
    HMDPositionLocal["fp_4"]["z"].append(frame['HMDPositionLocal']["z"])
    
for frame in FP_tutorial:
    timestamps["fp_tutorial"].append(frame['TimeStamp'])
    HMDPositionGlobal["fp_tutorial"]["x"].append(frame['HMDPositionGlobal']["x"])
    HMDPositionGlobal["fp_tutorial"]["y"].append(frame['HMDPositionGlobal']["y"])
    HMDPositionGlobal["fp_tutorial"]["z"].append(frame['HMDPositionGlobal']["z"])
    HMDPositionLocal["fp_tutorial"]["x"].append(frame['HMDPositionLocal']["x"])
    HMDPositionLocal["fp_tutorial"]["y"].append(frame['HMDPositionLocal']["y"])
    HMDPositionLocal["fp_tutorial"]["z"].append(frame['HMDPositionLocal']["z"])
    
for frame in Hybrid_area_1:
    timestamps["hybrid_1"].append(frame['TimeStamp'])
    HMDPositionGlobal["hybrid_1"]["x"].append(frame['HMDPositionGlobal']["x"])
    HMDPositionGlobal["hybrid_1"]["y"].append(frame['HMDPositionGlobal']["y"])
    HMDPositionGlobal["hybrid_1"]["z"].append(frame['HMDPositionGlobal']["z"])
    HMDPositionLocal["hybrid_1"]["x"].append(frame['HMDPositionLocal']["x"])
    HMDPositionLocal["hybrid_1"]["y"].append(frame['HMDPositionLocal']["y"])
    HMDPositionLocal["hybrid_1"]["z"].append(frame['HMDPositionLocal']["z"])
    
for frame in Hybrid_area_2:
    timestamps["hybrid_2"].append(frame['TimeStamp'])
    HMDPositionGlobal["hybrid_2"]["x"].append(frame['HMDPositionGlobal']["x"])
    HMDPositionGlobal["hybrid_2"]["y"].append(frame['HMDPositionGlobal']["y"])
    HMDPositionGlobal["hybrid_2"]["z"].append(frame['HMDPositionGlobal']["z"])
    HMDPositionLocal["hybrid_2"]["x"].append(frame['HMDPositionLocal']["x"])
    HMDPositionLocal["hybrid_2"]["y"].append(frame['HMDPositionLocal']["y"])
    HMDPositionLocal["hybrid_2"]["z"].append(frame['HMDPositionLocal']["z"])

for frame in Hybrid_area_3:
    timestamps["hybrid_3"].append(frame['TimeStamp'])
    HMDPositionGlobal["hybrid_3"]["x"].append(frame['HMDPositionGlobal']["x"])
    HMDPositionGlobal["hybrid_3"]["y"].append(frame['HMDPositionGlobal']["y"])
    HMDPositionGlobal["hybrid_3"]["z"].append(frame['HMDPositionGlobal']["z"])
    HMDPositionLocal["hybrid_3"]["x"].append(frame['HMDPositionLocal']["x"])
    HMDPositionLocal["hybrid_3"]["y"].append(frame['HMDPositionLocal']["y"])
    HMDPositionLocal["hybrid_3"]["z"].append(frame['HMDPositionLocal']["z"])

for frame in Hybrid_area_4:
    timestamps["hybrid_4"].append(frame['TimeStamp'])
    HMDPositionGlobal["hybrid_4"]["x"].append(frame['HMDPositionGlobal']["x"])
    HMDPositionGlobal["hybrid_4"]["y"].append(frame['HMDPositionGlobal']["y"])
    HMDPositionGlobal["hybrid_4"]["z"].append(frame['HMDPositionGlobal']["z"])
    HMDPositionLocal["hybrid_4"]["x"].append(frame['HMDPositionLocal']["x"])
    HMDPositionLocal["hybrid_4"]["y"].append(frame['HMDPositionLocal']["y"])
    HMDPositionLocal["hybrid_4"]["z"].append(frame['HMDPositionLocal']["z"])


currentkeys = list(timestamps.keys())
```

```python
sns.set()
fig, ax = plt.subplots(nrows=2, ncols=5, figsize=(15,5))
fig.suptitle("Timestamps within conditions", fontweight='bold', y=1.02)
plt.subplots_adjust(wspace=0.3, hspace=1)
#plt.tight_layout()
ax = ax.ravel()

i = 0
for key in currentkeys:
    if timestamps[key]:
        ax[i].plot(timestamps[key])
        ax[i].set_title(key, pad=18, fontweight='bold')
        i += 1
```

```python
sns.set()
fig, ax = plt.subplots(nrows=2, ncols=5, figsize=(15,5))
fig.suptitle("Global position X and Z", fontweight='bold', y=1.02)
plt.subplots_adjust(wspace=0.5, hspace=1)
ax = ax.ravel()

i = 0
for key in currentkeys:
    if HMDPositionGlobal[key]["x"]:
        ax[i].plot(HMDPositionGlobal[key]["x"], HMDPositionGlobal[key]["z"])
        ax[i].set_title(key, pad=18, fontweight='bold')
        start, end = ax[i].get_xlim()
        ax[i].xaxis.set_ticks(np.around(np.arange(start, end, abs(start-end)/5)))
        start, end = ax[i].get_ylim()
        ax[i].yaxis.set_ticks(np.around(np.arange(start, end, abs(start-end)/5)))
        ax[i].set_xlabel("x position")
        ax[i].set_ylabel("z position")
        i += 1
```
