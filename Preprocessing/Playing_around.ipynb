```python
sns.set()
fig, ax = plt.subplots(nrows=2, ncols=5, figsize=(15,5))
fig.suptitle("Global position X and Z", fontweight='bold', y=1.02)
plt.subplots_adjust(wspace=0.5, hspace=1)
#plt.tight_layout()
ax = ax.ravel()

i = 0
for key in currentkeys:
    if HMDPositionGlobalX[key]:
        ax[i].plot(HMDPositionGlobalX[key], HMDPositionGlobalZ[key])
        ax[i].set_title(key, pad=18, fontweight='bold')
        start, end = ax[i].get_xlim()
        ax[i].xaxis.set_ticks(np.around(np.arange(start, end, abs(start-end)/5)))
        start, end = ax[i].get_ylim()
        ax[i].yaxis.set_ticks(np.around(np.arange(start, end, abs(start-end)/5)))
        ax[i].set_xlabel("x position")
        ax[i].set_ylabel("z position")
        i += 1
```
